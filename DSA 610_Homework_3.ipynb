```python
# Create a dictionary to define the grading criteria
grading_criteria = {
    'A': lambda grade: grade > 90,
    'B': lambda grade: 80 <= grade < 90,
    'C': lambda grade: 70 <= grade < 80,
    'Did not pass': lambda grade: grade < 70
}

# Function to assign grades based on the grading criteria
def assign_grade(grade):
    for key, condition in grading_criteria.items():
        if condition(grade):
            return key

# Convert the "Grade" column to numeric
df1['Grade'] = pd.to_numeric(df1['Grade'], errors='coerce')

# Add the "grades" column to the DataFrame
df1['Grades'] = df1['Grade'].apply(assign_grade)

# Print the updated DataFrame
print(df1)
```

```python
#QUES 3.	How many students received A’s (above 90%), B’s (between 80% and less than 90%), 
# C’s (between 70% and less than 80%), or did not pass (less than 70%)?

# Count the number of students in each grade category
grade_counts = df1['Grades'].value_counts()
# Access the count for each grade category
a_count = grade_counts.get('A', 0)
b_count = grade_counts.get('B', 0)
c_count = grade_counts.get('C', 0)
did_not_pass_count = grade_counts.get('Did not pass', 0)
# Print the results
print("Number of students who received:")
print("A's          (above 90%)           :", a_count)
print("B's (between 80% and less than 90%):", b_count)
print("C's (between 70% and less than 80%):", c_count)
print("Did not pass (less than 70%)       :", did_not_pass_count)
```
